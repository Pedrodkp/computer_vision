```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('course material/DATA/00-puppy.jpg')
plt.imshow(img)
img_orig = img.copy()
```

```python
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
img = cv2.cvtColor(img,cv2.COLOR_RGB2HSV)
plt.imshow(img)
```

```python
img = img_orig.copy()
img = cv2.cvtColor(img,cv2.COLOR_BGR2HSV)
plt.imshow(img)
```
