```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
flat_chess = cv2.imread('course material/DATA/flat_chessboard.png')
plt.imshow(flat_chess)
#the grid should be specific made to looks like a chessboard
```

```python
found, corners = cv2.findChessboardCorners(flat_chess, (7,7)) #8x8 will not find the edge corners
```

```python
found
```

```python
corners
```

```python
cv2.drawChessboardCorners(flat_chess, (7,7), corners, found)
```

```python
plt.imshow(flat_chess)
```

```python
dots = cv2.imread('course material/DATA/dot_grid.png')
plt.imshow(dots)
```

```python
found, corners = cv2.findCirclesGrid(dots, (10,10), cv2.CALIB_CB_SYMMETRIC_GRID)
```

```python
found
```

```python
cv2.drawChessboardCorners(dots, (10,10), corners, found)
plt.imshow(dots)
```
