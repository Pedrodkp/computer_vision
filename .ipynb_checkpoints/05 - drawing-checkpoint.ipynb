```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
blank_img = np.zeros(shape=(512,512,3), dtype=np.int16) #dtype=data type off the array
```

```python
blank_img.shape
```

```python
plt.imshow(blank_img)
```

```python
#pt1 is top/left corner and pt2 is bottom/right corner
#thickness is border, negative value fill the rectangle
cv2.rectangle(blank_img,pt1=(384,20),pt2=(490,150),color=(0,255,0),thickness=5) 
```

```python
plt.imshow(blank_img)
```

```python
cv2.rectangle(blank_img,pt1=(200,200),pt2=(300,300),color=(0,0,255),thickness=5)
plt.imshow(blank_img)
```

```python
cv2.circle(img=blank_img,center=(100,100),radius=50,color=(255,0,0), thickness=8)
plt.imshow(blank_img)
```

```python
cv2.circle(img=blank_img,center=(400,400),radius=50,color=(255,0,0), thickness=-1)
plt.imshow(blank_img)
```

```python
cv2.line(blank_img,pt1=(0,0),pt2=(512,512),color=(102,255,255),thickness=5)
plt.imshow(blank_img)
```

```python
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(blank_img,text='Hello',org=(10,500),fontFace=font,fontScale=4,color=(255,255,255),thickness=3,lineType=cv2.LINE_AA) #org X,Y left/bottom start
plt.imshow(blank_img)
```

```python
blank_img = np.zeros(shape=(512,512,3),dtype=np.int32)
plt.imshow(blank_img)
```

```python
vertices = np.array([ [100,300], [200,200], [400,300], [200,400] ], dtype=np.int32)
```

```python
vertices
```

```python
vertices.shape
```

```python
pts = vertices.reshape((-1,1,2))
```

```python
pts.shape
```

```python
pts
```

```python
cv2.polylines(blank_img,[pts],isClosed=True,color=(255,0,0), thickness=5)
plt.imshow(blank_img)
```
