```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def display_img(img, cmap='gray'):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap)
```

```python
sep_coins = cv2.imread('course material/DATA/pennies.jpg')
display_img(sep_coins)
```

```python
# 1 - median blur
sep_blur = cv2.medianBlur(sep_coins,25)
display_img(sep_blur)
```

```python
# 2 - grayscale
gray_sep_coins = cv2.cvtColor(sep_blur, cv2.COLOR_BGR2GRAY)
display_img(gray_sep_coins)
```

```python
# 3 - binary threshold
ret, sep_thresh = cv2.threshold(gray_sep_coins, 160, 255, cv2.THRESH_BINARY_INV) #you can use the common value like 127 (middle of color) but will get noise, so, adjuste time by time
display_img(sep_thresh)
```

```python
# 4 - find contours 
contours, hierarchy = cv2.findContours(sep_thresh.copy(), cv2.RETR_CCOMP, cv2.CHAIN_APPROX_SIMPLE)
```

```python
for i in range(len(contours)):
    if hierarchy[0][i][3] == -1: #set -1 to get the internal contours
        cv2.drawContours(sep_coins, contours, i, (255,0,0), 10)
```

```python
display_img(sep_coins)
```

```python
img = cv2.imread('course material/DATA/pennies.jpg')
```

```python
img = cv2.medianBlur(img, 35) #use more blur for big images
```

```python
display_img(img)
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
ret, thresh = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY_INV)
```

```python
display_img(thresh) #the blur not remove all the resolution so we still get noise
```

```python
#https://en.wikipedia.org/wiki/Otsu%27s_method
ret, thresh = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY_INV+cv2.THRESH_OTSU)
```

```python
display_img(thresh)
```

```python
# noise removal (optional)
kernel = np.ones((3,3), np.uint8)
kernel
```

```python
opening = cv2.morphologyEx(thresh, cv2.MORPH_OPEN, kernel, iterations=2)
display_img(opening)
```

```python
sure_bg = cv2.dilate(opening,kernel,iterations=3)
display_img(sure_bg)
```

```python
#https://en.wikipedia.org/wiki/Distance_transform
dist_transform = cv2.distanceTransform(opening, cv2.DIST_L2, 5)
```

```python
display_img(dist_transform)
```

```python
ret, sure_fg = cv2.threshold(dist_transform, 0.7*dist_transform.max(), 255, 0)
display_img(sure_fg)
```

```python
sure_fg = np.uint8(sure_fg)
unknown = cv2.subtract(sure_bg, sure_fg)
display_img(unknown) #this is the region the ia not sure what is foreground or background
```

```python
ret, markers = cv2.connectedComponents(sure_fg)
markers = markers + 1
markers[unknown==255] = 0
display_img(markers)
```
