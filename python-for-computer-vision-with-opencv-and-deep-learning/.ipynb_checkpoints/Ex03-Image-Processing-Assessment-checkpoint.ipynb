```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
def display_img(img,cmap=None):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap)
```

**TASK: Open and display the giraffes.jpg image that is located in the DATA folder.**

```python
giraffes = cv2.imread('course material/DATA/giraffes.jpg')
giraffes = cv2.cvtColor(giraffes, cv2.COLOR_BGR2RGB)
```

```python
display_img(giraffes)
```

**TASK:Apply a binary threshold onto the image.**

```python
giraffes = cv2.imread('course material/DATA/giraffes.jpg', 0) #ZERO for gray scales
ret, th1 = cv2.threshold(giraffes,127 ,255,cv2.THRESH_BINARY)
```

```python
display_img(th1,'gray')
```

**TASK: Open the giraffes.jpg file from the DATA folder and convert its colorspace to  HSV and display the image.**

```python
giraffes = cv2.imread('course material/DATA/giraffes.jpg')
hsv = cv2.cvtColor(giraffes, cv2.COLOR_BGR2HSV)
```

```python
display_img(hsv)
```

**TASK: Create a low pass filter with a 4 by 4 Kernel filled with values of 1/10 (0.01) and then use 2-D Convolution to blur the giraffer image (displayed in normal RGB)**

```python
kernel = np.ones(shape=(4,4),dtype=np.float32)/10
```

```python
kernel
```

```python
giraffes = cv2.imread('course material/DATA/giraffes.jpg')
giraffes = cv2.cvtColor(giraffes, cv2.COLOR_BGR2RGB)
dst = cv2.filter2D(src=giraffes, ddepth=-1, kernel=kernel)
```

```python
display_img(dst)
```

**TASK: Create a Horizontal Sobel Filter (sobelx from our lecture) with a kernel size of 5 to the grayscale version of the giraffes image and then display the resulting gradient filtered version of the image.**

```python
giraffes = cv2.imread('course material/DATA/giraffes.jpg',0)
```

```python
sobelx = cv2.Sobel(src=giraffes, ddepth=cv2.CV_64F, dx=1, dy=0, ksize=5)
```

```python
display_img(sobelx,'gray')
```

**TASK: Plot the color histograms for the RED, BLUE, and GREEN channel of the giaraffe image. Pay careful attention to the ordering of the channels.**

```python
giraffes = cv2.imread('course material/DATA/giraffes.jpg')
color = ('b','g','r')
for i, col in enumerate(color):
    histr = cv2.calcHist([giraffes],[i],None,[256],[0,256])
    plt.plot(histr, color=col)
    plt.xlim(0,256)
plt.title('historigram for giraffes')
```
