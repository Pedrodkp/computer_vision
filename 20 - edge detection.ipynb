```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('course material/DATA/sammy_face.jpg') 
#for edges we dont care about colors rgb or bgr or whanever, so no conversion need
plt.imshow(img)
```

```python
edges = cv2.Canny(image=img, threshold1=127, threshold2=127)
plt.imshow(edges)
```

```python
edges = cv2.Canny(image=img, threshold1=0, threshold2=255)
plt.imshow(edges)
```

```python
med_value = np.median(img)
med_value
```

```python
#lower threshould to either ZERO or 70% of the median value whichever is greater
lower = int(max(0,0.7*med_value))
# upper threshold to either 130% of the median or the max 255, whichever is smaller
upper = int(min(255,1.3*med_value))
```

```python
edges = cv2.Canny(image=img, threshold1=lower, threshold2=upper)
plt.imshow(edges)
```

```python
blurred_img = cv2.blur(img, ksize=(5,5))
edges = cv2.Canny(image=blurred_img, threshold1=lower, threshold2=upper)
plt.imshow(edges)
```

```python
blurred_img = cv2.blur(img, ksize=(5,5)) #we can increase or decrease values to better performance
edges = cv2.Canny(image=blurred_img, threshold1=lower, threshold2=upper+50) #we can increase or decrease values to better performance
plt.imshow(edges)
```
