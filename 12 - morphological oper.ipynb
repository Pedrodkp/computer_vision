```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def load_img():
    blank_img = np.zeros((600,600))
    font = cv2.FONT_HERSHEY_SIMPLEX
    cv2.putText(blank_img, text='ABCDE', org=(50,300), fontFace=font, fontScale=5, color=(255,255,255), thickness=25
            )
    return blank_img
```

```python
def display_img(img):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap='gray')
```

```python
img = load_img()
display_img(img)
```

```python
kernel = np.ones((5,5), dtype=np.uint8) #this kernel is to set the background, 1 here is black
kernel
```

```python
result = cv2.erode(img, kernel, iterations=1) #erode (erosão) try to remove the thickness
display_img(result)
```

```python
result = cv2.erode(img, kernel, iterations=4) #more interations and 
display_img(result)
```

```python
#restart to make the oposite of erode, 
img = load_img()
display_img(img)
```

```python
white_noise = np.random.randint(low=0,high=2,size=(600,600)) #noise here goes from ZERO-1, because the imagem is black and white
white_noise
```

```python
display_img(white_noise)
```

```python
img.max() #image is black and white, but not ZERO-1
```

```python
white_noise = white_noise * 255# now 1 will became 255
```

```python
noise_img = white_noise + img
display_img(noise_img) #now we have the noise
```

```python
opening = cv2.morphologyEx(noise_img, cv2.MORPH_OPEN, kernel) #opening almost clear the noise
display_img(opening)
```

```python
display_img(img)
```

```python
img = load_img()
black_noise = np.random.randint(low=0,high=2,size=(600,600))
black_noise = black_noise * -255
black_noise
```

```python
black_noise_img = img + black_noise #we just make this step to show de ZEROS from image, and -255 added by black_noise, which will became ZEROS
black_noise_img
```

```python
black_noise_img[black_noise_img == -255] = 0
black_noise_img
```

```python
display_img(black_noise_img) #we can create a nice effect this way, because the noise was removed only from the background
```

```python
closing = cv2.morphologyEx(black_noise_img, cv2.MORPH_CLOSE, kernel) #now we can remove the last noise with the kernel
display_img(closing)
```
